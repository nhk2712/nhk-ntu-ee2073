#### Question 2. Audio amplifier characterization

Import required libraries

```python
import numpy as np
import plotly.graph_objs as go
```

Create NumPy arrays

```python
v_a = np.array([5.4, 4.5, 3.6, 2.7, 1.8, 0.9, 0.0, -0.9])
v_signal_pp = np.array([8.07, 6.37, 4.35, 2.61, 1.73, 774.37e-3, 218.36e-3, 126.29e-3])
v_audio_pp = np.array([1.29, 2.49, 4.27, 6.29, 10.00, 10.15, 7.55, 8.89])
v_volume_rms = np.array([95.04e-3, 413.14e-3, 947.54e-3, 1.72, 3.07, 3.16, 2.17, 2.96])
```

Calculate the required values

```python
v_c = 3.3*v_a/(56+3.3)
gain_db = 20*np.log10(v_audio_pp/v_signal_pp)
```

Plot Gain (dB) vs V<sub>C</sub> (V)

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=v_c, y=gain_db, mode='lines+markers'))
fig.update_layout(xaxis_title='V_C (V)', yaxis_title='Gain (dB)')
```

Comment: The plotted line is almost linear, which is the same as the plot in VCA lab. The line is shifted above for 30dB due to the appearance of the PA. Therefore, the audio amplifier system works properly as expected.
