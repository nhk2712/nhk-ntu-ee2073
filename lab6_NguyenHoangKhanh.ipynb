#### Question 2. PA frequency response characterization

Import required libraries:

```python
import numpy as np
import plotly.graph_objs as go
import math
```

Data in Python for Set 1 (Amplitude = 0.1 V):

```python
vin_1 = np.array([244.20e-3, 256.33e-3, 266.00e-3, 209.32e-3, 209.47e-3, 238.34e-3, 233.06e-3, 222.89e-3, 221.36e-3])
vout_1 = np.array([7.45, 8.70, 9.42, 8.32, 8.29, 8.64, 8.61, 8.72, 7.09])
```

Calculate gain (dB) for Set 1:

```python
gain_db_1 = [0]*len(vin_1)
for i in range(len(vin_1)):
    gain_db_1[i] = 20*math.log(vout_1[i]/vin_1[i], 10)
```

```python
gain_db_1
```

Data for Set 2 (Amplitude = 0.2 V):

```python
vin_2 = [441.04e-3, 437.60e-3, 467.17e-3, 409.06e-3, 404.38e-3, 413.64e-3, 395.64e-3, 420.40e-3, 410.60e-3]
vout_2 = [11.65, 11.05, 10.35, 10.13, 9.96, 10.09, 10.13, 10.10, 9.57]
```

Calculate gain (dB) for Set 2:

```python
gain_db_2 = [0]*len(vin_2)
for i in range(len(vin_2)):
    gain_db_2[i] = 20*math.log(vout_2[i]/vin_2[i], 10)
```

```python
gain_db_2
```

Plot gain (dB) vs frequency (Hz)

```python
freq = [100, 200, 500, 1000, 2000, 5000, 10000, 20000, 50000]

fig = go.Figure()
fig.update_xaxes(type='log')
fig.add_trace(go.Scatter(x=freq, y=gain_db_1, mode='lines+markers', name='Set 1'))
fig.add_trace(go.Scatter(x=freq, y=gain_db_2, mode='lines+markers', name='Set 2'))
fig.update_layout(xaxis_title='Frequency (Hz)', yaxis_title='Gain (dB)')
```

Comment: Set 1 (0.1Vpk input) yields higher gain vs frequency than Set 2 (0.2Vpk input). The overall trend fluctuates for both sets.
