#### Question 2. VU meter gain characterization

Import required libraries

```python
import numpy as np
import plotly.graph_objs as go
```

Create NumPy arrays

```python
v_IN_pp = np.array([8.01, 6.35, 6.18, 5.55, 4.22, 3.52, 1.89, 0.38969, 0.58488])
v_OUT1_pp = np.array([3.20, 2.98, 2.99, 2.70, 2.43, 1.90, 1.37, 0.61321, 0.71003])
v_OUT2_pp = np.array([2.37, 2.16, 2.19, 2.04, 1.49, 1.25, 0.61228, 0.13681, 0.19565])
v_OUT3_rms = np.array([1.81, 1.58, 1.53, 1.27, 0.95053, 0.65651, 0.22177, 0.01002, 0.00669])
```

Calculate the gain

```python
gain = 2*v_OUT2_pp/v_IN_pp
```

```python
gain
```

Plot gain vs V_IN

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=v_IN_pp, y=gain, mode='lines+markers', name='Gain'))
fig.update_layout(xaxis_title='V_IN_pp (Vpp)', yaxis_title='Gain (linear)')
```

Plot V_OUT3_rms vs V_IN

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=v_IN_pp, y=v_OUT3_rms, mode='lines+markers', name='V_OUT3_rms'))
fig.update_layout(xaxis_title='V_IN_pp (Vpp)', yaxis_title='V_OUT3_rms (V)')
```
